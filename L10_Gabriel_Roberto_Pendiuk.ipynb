```python
from matplotlib import pyplot as plt
from skimage.color import rgb2gray
from skimage import io, filters, data, util
import numpy as np
```

```python
img = io.imread('linkedin.jpg')
grayPic = rgb2gray(img)
grayPic = util.img_as_ubyte(grayPic)
```

```python
lin, col = np.shape(grayPic)
zerada = np.ones((lin,col))
t1 = filters.threshold_otsu(grayPic)
seg1 = np.where(grayPic>t1, 1, 0)
t2 = filters.rank.otsu(grayPic, zerada)
seg2 = np.where(grayPic>t2, 1, 0)
```

```python
plt.figure(figsize=(20,20))
plt.subplot(131)
plt.imshow(grayPic, cmap='gray')
plt.subplot(132)
plt.imshow(seg1, cmap='gray')
plt.subplot(133)
plt.imshow(seg2, cmap='gray')
```

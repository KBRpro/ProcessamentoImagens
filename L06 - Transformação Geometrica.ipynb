```python
from skimage import io, transform
import numpy as np
import matplotlib.pyplot as plt
```

```python
arq = 'dourado.jpeg'
c = [[200,188], [600,188], [100,600], [600,600]]
linha = 550
```

```python

# Load source image
img = io.imread(arq)
img = np.fliplr(img)
```

```python
plt.figure()
# Adjust the 4 coordinates below: they must lie on the same plane. Use pitch patterns to axis-align.
#plt.rcParams["figure.figsize"] = (20, 20)
plt.subplot(121)
plt.imshow(img)

```

```python
pts = np.array(c)
```

```python
plt.scatter(pts[:,0], pts[:,1])
```

```python
pts_norm = np.array([[0,0], [400,0], [0, 800], [400, 800]])
# Calculate Homography
tform = transform.ProjectiveTransform()
tform.estimate(pts_norm, pts)
```

```python
## Straighten out image
img_norm =  transform.warp(img, tform, output_shape=(800, 400))
img_norm
```

```python
plt.subplot(121)
plt.imshow(img_norm)
x1= [115,115]
x2= [800,0]
plt.plot(x1, x2, color="white", linewidth=3)
```

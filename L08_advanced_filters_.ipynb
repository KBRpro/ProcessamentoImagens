```python
from skimage import io, data
import scipy.signal
import numpy as np
import cv2
from skimage import data, color, filters
from skimage.transform import rescale, resize, downscale_local_mean
from matplotlib import pyplot as plt3
from skimage.filters import gaussian
from skimage.filters import meijering
from skimage.filters import sato
from skimage.filters import threshold_isodata
from skimage.filters import threshold_niblack
from skimage.filters import laplace
```

```python
im= plt.imread('kbr.jpg')
im= im.astype('float')
im /= 255.
```

```python
imout=gaussian(im, sigma=0.4)
plt.imshow(imout) #suaviza a imagem(passa baixa)
```

```python
imout=meijering(im, sigmas=range(1, 10, 2), alpha=None, black_ridges=True)
```

```python
plt.imshow(imout) #invertte as cores e detecta partes continuas de neuritos
#passa alta
```

```python
imout=sato(imout, sigmas=range(1, 10, 2), black_ridges=True)
```

```python
imout=laplace(im, ksize=3, mask=None)
plt.imshow(imout)
#filtro passa alto
#retorna o mapa de edges
```

```python
threshold_isodata(image=im, nbins=256, return_all=False)
```

```python
#mostra quantos thereshold se encontram na imagem
```

```python
q = 1
threshold_image = threshold_niblack(im, k=0) * q
plt.imshow(threshold_image)
#filtro passa alto
#retorna a imagem em branco e preto com os pixels usando a formula
#T = m(x,y) - k * s(x,y)
```
